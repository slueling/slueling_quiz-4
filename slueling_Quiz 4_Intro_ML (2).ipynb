```python
from sklearn.datasets import load_iris
from sklearn import tree
```

```python
iris = load_iris()
X,y = iris.data, iris.target
```

```python
X.shape, y.shape
```

```python
clf = tree.DecisionTreeClassifier(criterion = 'entropy', random_state = 42, max_depth=3)
clf = clf.fit(X, y)
```

```python
clf.score(X, y) 
round(clf.score(X, y) , 3)
```
